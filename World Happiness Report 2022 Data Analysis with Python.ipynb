# **World Happiness Report 2022 Data Analysis with Python**
## **About Dataset**
The first World Happiness Report was released on April 1, 2012 as a foundational text for the UN High Level Meeting: Well-being and Happiness: Defining a New Economic Paradigm, drawing international attention. The report outlined the state of world happiness, causes of happiness and misery, and policy implications highlighted by case studies. In 2013, the second World Happiness Report was issued, and since then has been issued on an annual basis with the exception of 2014. The report primarily uses data from the Gallup World Poll. Each annual report is available to the public to download on the World Happiness Report website.

## **Content**
The happiness scores and rankings use data from the Gallup World Poll. The scores are based on answers to the main life evaluation question asked in the poll. This question, known as the Cantril ladder, asks respondents to think of a ladder with the best possible life for them being a 10 and the worst possible life being a 0 and to rate their own current lives on that scale. The scores are from nationally representative samples for the years 2013-2022 and use the Gallup weights to make the estimates representative. The columns following the happiness score estimate the extent to which each of six factors – economic production, social support, life expectancy, freedom, absence of corruption, and generosity.

The following columns: GDP per Capita, Family, Life Expectancy, Freedom, Generosity, Government Corruption describe the extent to which these factors contribute in evaluating the happiness in each country.

### 4. Data Visualization
We can understand the dataset much much better with visualization. Now, let look at the relationships of numeric variables using the seaborn. First, let's import the tools we will need.

```python
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
import plotly.graph_objects as go
```

In order to check the distribution of values within a column, we can employ various plots, and one of the best ways to easily display the distribution of values is a histogram.

```python
sns.displot(data=data2022, x="Happiness score", label='2022', kde = True, color="blue")
plt.title('Histogram of Happiness', fontsize=17)
```

**Relationships between Variables**

Now that we have a good idea about the structure of the dataset, we can dive into relationships between variables. One of the ways to quickly check the correlations between variables is by running df.corr() which we did ealier, which returned a table of all correlations. However, to get a more clear view of correlations, we turn to data visualization and one of the best method is a heatmap.

```python
sns.heatmap(data2022.corr(), annot= True)
```

The heatmap produces darker colors for stronger correlations, and the annotations show the correlation coefficient. For example, ‘Healthy Life Expectancy’ and ‘GDP per capita’ have a 0.82 correlation, which means these two variables are strongly correlated.

Another great way to take a look at the relationship between two variables is a scatter plot. Note, here we are only looking at two variables at a time. A scatter plot is a very useful tool on its own, but they also complement the correlation matrix very well, helping you draw more insights from the data.

```python
plt.scatter(data = data2022, x = "GDP per capita", y = "Happiness score")
plt.xlabel('GDP per capita')
plt.ylabel('Happiness Score')
plt.title('GDP vs. Happiness', fontsize=17)
```

We can see a strong correlation between Gross Domestic Product (GDP) and Happiness.

```python
plt.scatter(data = data2022, x = "Healthy Life Expectancy", y = "GDP per capita")
plt.xlabel('Healthy Life Expectancy')
plt.ylabel('GDP per capita')
plt.title('GDP per capita vs. Healthy Life Expectancy', fontsize=15)
```

Also, a strong correlation between GDP and Healthy Life Expectancy, wealthier countries live a longer healthier life.

```python
plt.scatter(data = data2022, x = "Healthy Life Expectancy", y = "Happiness score")
plt.xlabel('Healthy Life Expectancy')
plt.ylabel('Happiness Score')
plt.title('Healthy Life Expectancy vs. Happiness', fontsize=17)
```

Positive correlation between Healthy Life Expectancy and Happiness.

```python
plt.scatter(data = data2022, x = "Freedom to make life choices", y = "Happiness score")
plt.xlabel('Freedom')
plt.ylabel('Happiness Score')
plt.title('Freedom vs. Happiness', fontsize=17)
```

Countries with more freedom to make life choices also positively correlates to happiness.

```python
plt.scatter(data = data2022, x = "Social Support", y = "Happiness score")
plt.xlabel('Social Support')
plt.ylabel('Happiness Score')
plt.title('Social Support vs. Happiness', fontsize=17)
```

As well as Social Support.
